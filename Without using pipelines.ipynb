```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
# filling null values

si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

x_train_age = si_age.fit_transform(x_train[['Age']])
x_train_embarked = si_embarked.fit_transform(x_train[['Embarked']])

x_test_age = si_age.fit_transform(x_test[['Age']])
x_test_embarked = si_embarked.fit_transform(x_test[['Embarked']])
```

```python
x_test_age
```

```python
x_train.isnull().sum()
```

```python
# ohe on sex and embarked

ohe_sex = OneHotEncoder(sparse=False,handle_unknown='ignore',drop='first')
ohe_embarked = OneHotEncoder(sparse=False,handle_unknown='ignore',drop='first')

x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked = ohe_embarked.fit_transform(x_train_embarked)

x_test_sex = ohe_sex.fit_transform(x_test[['Sex']])
x_test_embarked = ohe_embarked.fit_transform(x_test_embarked)
```

```python
x_test_sex
```

```python
x_train_embarked
```

```python
y_test
```

```python
x_train_rem = x_train.drop(columns = ['Sex','Age','Embarked'])
```

```python
x_test_rem = x_test.drop(columns = ['Sex','Age','Embarked'])
```

```python
x_train_rem.head(1)
```

```python
x_train_transformed = np.concatenate((x_train_rem,x_train_age,x_train_sex,x_train_embarked),axis=1)
x_test_transformed = np.concatenate((x_test_rem,x_test_age,x_test_sex,x_test_embarked),axis=1)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed,y_train)
```

```python
x_test_transformed.shape
```

```python
y_pred = clf.predict(x_test_transformed)
```
